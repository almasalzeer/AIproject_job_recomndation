# Job Skill Set Dataset - Initial Description
This dataset contains information about various job positions and their required skills. 
The goal is to analyze job market trends and identify key skills required for different roles.


```python
import pandas as pd  
import matplotlib.pyplot as plt 
import seaborn as sns  

sns.set(style="darkgrid")
```

## Dataset Information
- **Goal:** The dataset aims to provide insights into job market requirements and skills.
- **Purpose:** Collected to help job seekers, HR professionals, and researchers understand the most in-demand skills.
- **Source:** [Kaggle - Job Skill Set Dataset](https://www.kaggle.com/datasets/batuhanmutlu/job-skill-set)

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18,8))  
top_skills = df["job_skill_set"].value_counts().head(10)  

top_skills.plot(kind="bar", color="skyblue")  

plt.title("must needed skills", fontsize=16) 
plt.xlabel("skill", fontsize=14)  
plt.ylabel("appearance time", fontsize=14)  

plt.xticks(rotation=30, ha="right", fontsize=12)  

plt.grid(axis="y", linestyle="--", alpha=0.7) 
plt.show()
```

```python
df_exploded = df.explode("job_skill_set")  
category_skill_counts = df_exploded.groupby("category")["job_skill_set"].count().reset_index()

category_skill_counts = category_skill_counts.sort_values(by="job_skill_set", ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="job_skill_set", y="category", data=category_skill_counts)

plt.xlabel("Number of Required Skills")
plt.ylabel("Job Category")
plt.title("Number of Required Skills per Job Category")

plt.show()
```
