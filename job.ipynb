### Data Preprocessing

In this section, we perform several preprocessing steps to to ensure our dataset is cleaned, transformed, and ready for machine learning. Below is a breakdown of the steps we'll take:


1. **Handling Missing Values**
   We have already verified that there are no missing values in the dataset, so no imputation is needed. This ensures our data is complete and ready       for further processing.  

2. **Handling Duplicates**
   We have already checked that there are no duplicate rows in the dataset. Removing duplicates is crucial to prevent bias and redundancy in model         training, but since none are present, no modifications are required.  
    
3. **Text Cleaning** 
   Since our dataset includes textual data, we apply preprocessing techniques to improve model performance:

   - **Removing Numerical Values (from job titles, job descriptions, and skills):** To ensure meaningful text analysis, we remove all numerical values     from job titles, job descriptions, and skills a regular expression. Numbers often do not contribute relevant context in these fields, so eliminating    them helps maintain data clarity. The cleaned text is stored in new columns, preserving the original data for reference while ensuring the processed    data remains focused on relevant words and phrases.
   - **Lowercasing:** Convert all text to lowercase for consistency.  
   - **Removing Special Characters:** Eliminate unnecessary symbols, punctuation, and extra spaces while keeping meaningful symbols (e.g., hyphens in      "self-motivated" or programming terms like "C++").  
   - **Truncating Large Text:** In this step, we will refine the job descriptions by truncating the descriptions to a maximum of 2000 characters for       better readability. This ensures that the text is clear, relevant, and manageable for further analysis or display purposes.

5. **Discretization**
   Discretization is not necessary for our dataset, as it is typically used for transforming continuous numerical data into categorical bins. Our          dataset primarily consists of textual data, such as job titles, descriptions, and skills. Since there are no continuous numerical variables like        experience or salary, discretization does not apply at this stage. If we later incorporate attributes such as years of experience, we may consider      discretization to improve model interpretability.

6. **Normalization** 
   Normalization, which scales numerical values to a common range, is not required here because our dataset contains mainly text-based fields. For         instance, techniques like normalization would be relevant if we had features such as salary or experience that require scaling to avoid larger          values dominating model training. As there are no such numerical features currently in the dataset, normalization is unnecessary. We may reconsider     it if we add such attributes in the future.

7. **Variable Removal**
   We will first check if the `job_id` variable contains unique values. If confirmed, we will remove it as it does not contribute to the model’s           predictive power. The `job_id` is just a unique identifier with no meaningful relation to the job description, skills, or category, and including it    could introduce unnecessary complexity and biases. Removing it allows the model to focus on relevant features like job descriptions and skills,         improving its performance and generalizability.

```python
import pandas as pd
import re
import IPython.display as display

# Load the dataset 
df = pd.read_csv(r"C:\Users\Sara\Desktop\python_project\all_job_post.csv") 

# Function to remove numbers from text fields
def remove_numbers(text):
    return re.sub(r'\d+', '', str(text))  

# Function to remove special characters
def remove_special_characters(text):
    return re.sub(r'([^\w\s\-+.,])', '', text)  # Keeps hyphens, plus, periods, and commas

# Function to convert text to lowercase
def lowercase_text(text):
    return str(text).lower()

# Apply transformations directly on the existing dataset
df['job_title'] = df['job_title'].apply(lambda x: remove_numbers(str(x))) 
df['job_description'] = df['job_description'].apply(lambda x: remove_numbers(str(x)))
df['job_skill_set'] = df['job_skill_set'].apply(lambda x: remove_numbers(str(x)))
df['category'] = df['category'].apply(lambda x: remove_numbers(str(x))) 

df['job_title'] = df['job_title'].apply(lambda x: remove_special_characters(str(x)))  
df['job_description'] = df['job_description'].apply(lambda x: remove_special_characters(str(x)))
df['job_skill_set'] = df['job_skill_set'].apply(lambda x: remove_special_characters(str(x)))
df['category'] = df['category'].apply(lambda x: remove_special_characters(str(x)))  

df['job_title'] = df['job_title'].apply(lambda x: lowercase_text(str(x)))  
df['job_description'] = df['job_description'].apply(lambda x: lowercase_text(str(x)))
df['job_skill_set'] = df['job_skill_set'].apply(lambda x: lowercase_text(str(x)))
df['category'] = df['category'].apply(lambda x: lowercase_text(str(x))) 

# Display the cleaned DataFrame 
display.display(df[['job_id','category', 'job_title', 'job_description', 'job_skill_set']].head().style.set_table_attributes('class="table table-striped"'))

# Save the cleaned dataset
df.to_csv(r"C:\Users\Sara\Desktop\python_project\all_job_post.csv", index=False)
```
