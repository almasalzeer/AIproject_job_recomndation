```python
import pandas as pd

# Load the dataset from the specified file path
file_path = r'C:\Users\almas\Downloads\archive\all_job_post.csv'
df = pd.read_csv(file_path)

# Statistical summary for categorical columns (category, job_title, job_description, job_skill_set)
stat_summary = df.describe(include='object')  # Only consider object/categorical columns

# Create a table of meaningful statistical summary for categorical columns
stat_table = pd.DataFrame(stat_summary).transpose()

# Adding additional columns for frequency counts of the top values
for col in df.select_dtypes(include=['object']).columns:
    value_counts = df[col].value_counts()
    stat_table[col + ' (Top Value)'] = value_counts.idxmax()
    stat_table[col + ' (Frequency)'] = value_counts.max()

# Display the table in an easy-to-read format
import IPython.display as display
display.display(stat_table)

# Explanation 

# This table provides a statistical summary of the categorical columns in the dataset.
# We have four columns of interest: 'category', 'job_title', 'job_description', and 'job_skill_set'.

# 1. 'category':
# - There are 1167 entries in total, with 5 unique categories (such as 'HR', 'INFORMATION-TECHNOLOGY', etc.).
# - The most frequent category is 'INFORMATION-TECHNOLOGY', which appears 240 times.

# 2. 'job_title':
# - This column contains 1167 entries and 723 unique job titles.
# - The most frequent job title is 'Business Development Manager', appearing 53 times.

# 3. 'job_description':
# - There are 1167 job descriptions, with each one being unique, meaning there are no repeated descriptions in this dataset.
# - The top job description is: 'EXCITING OPPORTUNITY: GLOBAL BUSINESS DEVELOPMENT', which appears just once.

# 4. 'job_skill_set':
# - Like 'job_description', 'job_skill_set' has 1167 unique entries.
# - The most common skill set is ['strategic marketing', 'business development'], appearing once.

# Summary:
# - Most columns have unique values for each entry, with the exception of 'category' and 'job_title', where there are common entries.
# - The dataset contains diverse job titles and skill sets, with 'INFORMATION-TECHNOLOGY' being the most frequent category.
```

```python
from collections import Counter

# Combine all job skills into a single list
all_skills = ' '.join(df['job_skill_set'].dropna()).split(',')

# Count the frequency of each skill
skill_counts = Counter(all_skills)

# Get the top 10 most common skills
top_skills = dict(skill_counts.most_common(10))

# Plot the bar chart for job skills
plt.figure(figsize=(10, 6))
plt.bar(top_skills.keys(), top_skills.values(), color='salmon')
plt.title('Top 10 Most Common Job Skills')
plt.xlabel('Job Skills')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()

#Explanation 
#This chart shows the most common job skills required by employers based on the job postings.
#Insight: It identifies which skills are in demand across various industries, like "business development" or "marketing." It provides a deeper view of the qualifications employers are looking for.
```
